```python
def imgData(path):
    img = glymur.Jp2k(path)
    return {"path": path, "h": img.shape[0], "w": img.shape[1], "d": img.shape[2]}

photoinfo = list(map(imgData, g.glob('books/SovetskoePhoto_1967/*.jp2')))
list(photoinfo)
```

```python
def crops(pageinfo):
    longaxis = "w" if pageinfo["w"] > pageinfo["h"] else "h"
    length = pageinfo[longaxis]
    print(length)
```
